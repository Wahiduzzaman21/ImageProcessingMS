```python
import pickle
def unpickle(file):
    with open(file, 'rb') as fo:
        data = pickle.load(fo, encoding='bytes')
    return data
```

```python
# Eucledean Manhattan Calculation
import numpy as np
def calculateManhattanDistance(vector1, vector2):
    distance = np.sum(np.absolute(vector1-vector2))
    return distance
```

```python
# Eucledean Distance Calculation
import numpy as np
def calculateEucledeanDistance(vector1, vector2):
    distance = np.sqrt(np.sum((vector1-vector2)**2))
    return distance
```

```python
# Unpickle data file (This block of code, I have taken help from google. )
import numpy as np
cifar_train_data = None
cifar_train_labels = []
for i in range(1, 6):
    cifar_train_data_dict = unpickle("./cifar-10-batches-py" + "/data_batch_{}".format(i))
    if i == 1:
        cifar_train_data = cifar_train_data_dict[b'data']
    else:
        cifar_train_data = np.vstack((cifar_train_data, cifar_train_data_dict[b'data']))
    cifar_train_labels += cifar_train_data_dict[b'labels']
```

```python
# Convert the data shape from 3072 to 32*32*3
cifar_train_data = cifar_train_data.reshape((len(cifar_train_data), 3, 32, 32))
cifar_train_data = np.rollaxis(cifar_train_data, 1, 4)
```

```python
# categorize the data into different list
airplane_images= []
automobile_images = []
bird_images = []
cat_images =[]
deer_images = []
dog_images = []
frog_images = []
horse_images = []
ship_images = []
truck_images = []
for i in range (50000):
    if cifar_train_labels[i] == 0:
        airplane_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 1:
        automobile_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 2:
        bird_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 3:
        cat_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 4:
        deer_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 5:
        dog_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 6:
        frog_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 7:
        horse_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 8:
        ship_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 9:
        truck_images.append(cifar_train_data[i])
```

```python
# Convert all category list to np array
np_airplane = np.array(airplane_images) 
np_automobile = np.array(automobile_images) 
np_bird = np.array(bird_images) 
np_cat = np.array(cat_images) 
np_deer = np.array(deer_images) 
np_dog = np.array(dog_images) 
np_frog = np.array(frog_images) 
np_horse = np.array(horse_images) 
np_ship = np.array(ship_images) 
np_truck = np.array(truck_images) 

np_all = np.concatenate((np_airplane, np_automobile, np_bird, np_cat, np_deer, np_dog,
                        np_frog, np_horse, np_ship, np_truck), axis=0)
```

```python
random_number = np.random.randint(0, 5000)

print ("Please wait for Manhattan Distance calculation............................ :)")

manhattan_distance_airplane = [calculateManhattanDistance (np_airplane[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_airplane.sort()

manhattan_distance_automobile = [calculateManhattanDistance (np_automobile[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_automobile.sort()

manhattan_distance_bird = [calculateManhattanDistance (np_bird[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_bird.sort()

manhattan_distance_cat = [calculateManhattanDistance (np_cat[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_cat.sort()

manhattan_distance_deer = [calculateManhattanDistance (np_deer[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_deer.sort()

manhattan_distance_dog = [calculateManhattanDistance (np_dog[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_dog.sort()

manhattan_distance_frog = [calculateManhattanDistance (np_frog[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_frog.sort()

manhattan_distance_horse = [calculateManhattanDistance (np_horse[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_horse.sort()

manhattan_distance_ship = [calculateManhattanDistance (np_ship[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_ship.sort()

manhattan_distance_truck = [calculateManhattanDistance (np_truck[random_number], np_all[index]) for index in range(50000)]
manhattan_distance_truck.sort()

print ("Plotting the table..............")
import matplotlib.pyplot as plt
f, ax = plt.subplots(10, 6)
f.set_figheight(32)
f.set_figwidth(32)
for m in range(10):
        for n in range(6):
            if m == 0 and n==0:
                ax[m, n].imshow(np_airplane[random_number])
            elif m == 0 and n>0:
                ax[m,n].text(0.35, 0.45, str(manhattan_distance_airplane[n]), fontsize=20)
            
            elif m == 1 and n==0:
                ax[m, n].imshow(np_automobile[random_number])
            elif m == 1 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_automobile[n]), fontsize=20)
            
            elif m == 2 and n==0:
                ax[m, n].imshow(np_bird[random_number])
            elif m == 2 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_bird[n]), fontsize=20)

            elif m == 3 and n==0:
                ax[m, n].imshow(np_cat[random_number])
            elif m == 3 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_cat[n]), fontsize=20)
            
            elif m == 4 and n==0:
                ax[m, n].imshow(np_deer[random_number])
            elif m == 4 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_deer[n]), fontsize=20)

            elif m == 5 and n==0:
                ax[m, n].imshow(np_dog[random_number])
            elif m == 5 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_dog[n]), fontsize=20)
                
            elif m == 6 and n==0:
                ax[m, n].imshow(np_frog[random_number])
            elif m == 6 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_frog[n]), fontsize=20)
                
            elif m == 7 and n==0:
                ax[m, n].imshow(np_horse[random_number])
            elif m == 7 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_horse[n]), fontsize=20)
                
            elif m == 8 and n==0:
                ax[m, n].imshow(np_ship[random_number])
            elif m == 8 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_ship[n]), fontsize=20)
                
            elif m == 9 and n==0:
                ax[m, n].imshow(np_truck[random_number])
            elif m == 9 and n>0:
                ax[m, n].text(0.35, 0.45, str(manhattan_distance_truck[n]), fontsize=20)
            
            ax[m, n].get_xaxis().set_visible(False)
            ax[m, n].get_yaxis().set_visible(False)


print ("# Table 1- Top 5 lowest Manhattan distance of each category's random image with respect to all the images")

plt.show()
```
