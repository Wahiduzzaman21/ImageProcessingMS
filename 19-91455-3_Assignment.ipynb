```python
import pickle
def unpickle(file):
    with open(file, 'rb') as fo:
        data = pickle.load(fo, encoding='bytes')
    return data
```

```python
# Eucledean Manhattan Calculation
import numpy as np
def calculateManhattanDistance(vector1, vector2):
    distance = np.sum(np.absolute(vector1-vector2))
    return distance
```

```python
# Eucledean Distance Calculation
import numpy as np
def calculateEucledeanDistance(vector1, vector2):
    distance = np.sqrt(np.sum((vector1-vector2)**2))
    return distance
```

```python
# Unpickle data file (This block of code, I have taken help from google. )
import numpy as np
cifar_train_data = None
cifar_train_labels = []
for i in range(1, 6):
    cifar_train_data_dict = unpickle("./cifar-10-batches-py" + "/data_batch_{}".format(i))
    if i == 1:
        cifar_train_data = cifar_train_data_dict[b'data']
    else:
        cifar_train_data = np.vstack((cifar_train_data, cifar_train_data_dict[b'data']))
    cifar_train_labels += cifar_train_data_dict[b'labels']
```

```python
# Convert the data shape from 3072 to 32*32*3
cifar_train_data = cifar_train_data.reshape((len(cifar_train_data), 3, 32, 32))
cifar_train_data = np.rollaxis(cifar_train_data, 1, 4)

rgb_weights = [0.2125 , 0.5870, 0.1140]
cifar_train_data = np.dot(cifar_train_data[...,:3], rgb_weights)
```

```python
# categorize the data into different list
airplane_images= []
automobile_images = []
bird_images = []
cat_images =[]
deer_images = []
dog_images = []
frog_images = []
horse_images = []
ship_images = []
truck_images = []
for i in range (50000):
    if cifar_train_labels[i] == 0:
        airplane_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 1:
        automobile_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 2:
        bird_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 3:
        cat_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 4:
        deer_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 5:
        dog_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 6:
        frog_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 7:
        horse_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 8:
        ship_images.append(cifar_train_data[i])
    elif cifar_train_labels[i] == 9:
        truck_images.append(cifar_train_data[i])
```

```python
# Convert all category list to np array
np_airplane = np.array(airplane_images) 
np_automobile = np.array(automobile_images) 
np_bird = np.array(bird_images) 
np_cat = np.array(cat_images) 
np_deer = np.array(deer_images) 
np_dog = np.array(dog_images) 
np_frog = np.array(frog_images) 
np_horse = np.array(horse_images) 
np_ship = np.array(ship_images) 
np_truck = np.array(truck_images) 

np_all = np.concatenate((np_airplane, np_automobile, np_bird, np_cat, np_deer, np_dog,
                        np_frog, np_horse, np_ship, np_truck), axis=0)
```

```python
from numba import jit

@jit
def most_frequent(List): 
    counter = 0
    classLabel = -1
    for i in List: 
        curr_frequency = List.count(i) 
        if(curr_frequency> counter): 
            counter = curr_frequency 
            classLabel = i 
  
    return classLabel
```

```python
import numpy as np
from numba import jit

@jit
def predictClassKNN(testVector, trainDataSet, trainDatasetRange, trainLabel, numberOfK):
    eucledean_distance = [np.sqrt(np.sum((testVector-trainDataSet[index])**2)) for index in range(trainDatasetRange)]
    eucledean_distance_np = np.array(eucledean_distance)   
       
    sort_index = np.argsort(eucledean_distance_np)   
    
    nearest_labels = [trainLabel[indice] for indice in sort_index]
    nearest_labels = nearest_labels [0: numberOfK]
    
    predictedClass = most_frequent (nearest_labels)
    return predictedClass
    
```

```python
from numba import jit

@jit
def EachfoldValidation (testFold, testFoldLabels, testFoldSampleCount, trainFold, trainFoldLabels, 
                        trainFoldSampleCount, kNearest):


    prediction = [predictClassKNN(testFold[i],trainFold, trainFoldSampleCount, trainFoldLabels, kNearest)
                  for i in range (testFoldSampleCount)]
    
    counter = [prediction[i] for i in range (testFoldSampleCount) if prediction[i] == testFoldLabels[i]]
    
    foldAccuracy = (len(counter)/testFoldSampleCount) * 100
    
    return foldAccuracy
```

```python
import time
def kfoldValidation (imagedata, label, validationSetSamplesNo, trainsetSamplesNo, kNearest):
    
    cifar_train_data_portion_1 = imagedata[0:10000]
    cifar_train_labels_1 = label[0:10000]

    cifar_train_data_portion_2 = imagedata[10000:20000]
    cifar_train_labels_2 = label[10000:20000]

    cifar_train_data_portion_3 = imagedata[20000:30000]
    cifar_train_labels_3 = label[20000:30000]

    cifar_train_data_portion_4 = imagedata[30000:40000]
    cifar_train_labels_4 = label[30000:40000]

    cifar_train_data_portion_5 = imagedata[40000:50000]
    cifar_train_labels_5 = label[40000:50000]

    accuracyFold1=0
    accuracyFold2=0
    accuracyFold3=0
    accuracyFold4=0
    accuracyFold5=0
    
    for i in range (5):
        if i == 0:
            trainDataSet = np.concatenate((cifar_train_data_portion_2, cifar_train_data_portion_3, cifar_train_data_portion_4,
                               cifar_train_data_portion_5), axis=0)
            trainLabels = np.concatenate((cifar_train_labels_2, cifar_train_labels_3, cifar_train_labels_4,
                               cifar_train_labels_5), axis=0)
            accuracyFold1 = EachfoldValidation(cifar_train_data_portion_1, cifar_train_labels_1, validationSetSamplesNo, trainDataSet, trainLabels, trainsetSamplesNo, kNearest)
            
        elif i == 1:
            trainDataSet = np.concatenate((cifar_train_data_portion_1, cifar_train_data_portion_3, cifar_train_data_portion_4,
                               cifar_train_data_portion_5), axis=0)
            trainLabels = np.concatenate((cifar_train_labels_1, cifar_train_labels_3, cifar_train_labels_4,
                               cifar_train_labels_5), axis=0)
            accuracyFold2 = EachfoldValidation(cifar_train_data_portion_2, cifar_train_labels_2,
                                              validationSetSamplesNo, trainDataSet, trainLabels, trainsetSamplesNo, kNearest)
        elif i == 2:
            trainDataSet = np.concatenate((cifar_train_data_portion_1, cifar_train_data_portion_2, cifar_train_data_portion_4,
                               cifar_train_data_portion_5), axis=0)
            trainLabels = np.concatenate((cifar_train_labels_1, cifar_train_labels_2, cifar_train_labels_4,
                               cifar_train_labels_5), axis=0)
            accuracyFold3 = EachfoldValidation(cifar_train_data_portion_3, cifar_train_labels_3,
                                              validationSetSamplesNo, trainDataSet, trainLabels, trainsetSamplesNo, kNearest)
        
        elif i == 3:
            trainDataSet = np.concatenate((cifar_train_data_portion_1, cifar_train_data_portion_2, cifar_train_data_portion_3,
                               cifar_train_data_portion_5), axis=0)
            trainLabels = np.concatenate((cifar_train_labels_1, cifar_train_labels_2, cifar_train_labels_3,
                               cifar_train_labels_5), axis=0)
            accuracyFold4 = EachfoldValidation(cifar_train_data_portion_4, cifar_train_labels_4,
                                              validationSetSamplesNo, trainDataSet, trainLabels, trainsetSamplesNo, kNearest) 
        elif i == 4:
            trainDataSet = np.concatenate((cifar_train_data_portion_1, cifar_train_data_portion_2, cifar_train_data_portion_3,
                               cifar_train_data_portion_4), axis=0)
            trainLabels = np.concatenate((cifar_train_labels_1, cifar_train_labels_2, cifar_train_labels_3,
                               cifar_train_labels_4), axis=0)
            accuracyFold5 = EachfoldValidation(cifar_train_data_portion_5, cifar_train_labels_5,
                                              validationSetSamplesNo, trainDataSet, trainLabels, trainsetSamplesNo, kNearest)

    
    averageAccuracy  = (accuracyFold1+ accuracyFold2+ accuracyFold3+ accuracyFold4+ accuracyFold5) / 5
    Acc = [accuracyFold1,accuracyFold2,accuracyFold3,accuracyFold4,accuracyFold5]
    return averageAccuracy, Acc
```

```python
import matplotlib.pyplot as plt
import numpy as np

cifar_train_labels = np.array(cifar_train_labels)

testSamples = 10000
trainSamples = 40000

kneighbour = [1, 3, 5, 8, 10, 12, 15, 20, 50, 100]

kneighbourseachFold = [[i]*5 for i in kneighbour ]

Accurcies = []
mean_accuracy = []

for i in range (len(kneighbour)):
    acc = kfoldValidation (cifar_train_data, cifar_train_labels, testSamples, trainSamples, kneighbour[i])
    Accurcies.append(acc[1])  
    mean_accuracy.append(acc[0])



print ("# Task Three: Plot accuracy for each fold as well as mean accuracy against each K-Nearest Neighbour value")

accuracies_std = np.array([np.std(v) for v in sorted(Accurcies)])

plt.scatter(kneighbourseachFold, Accurcies)

plt.scatter(kneighbour, mean_accuracy, color ='orange')
plt.plot(kneighbour,mean_accuracy,'k-')

plt.errorbar(kneighbour, mean_accuracy, yerr=accuracies_std)

plt.title('Cross-validation on k')
plt.xlabel('K')
plt.ylabel('Cross validation accuracy(%)')

plt.show()
```

```python
cifar_test_data_dict = unpickle("./cifar-10-batches-py/test_batch")
cifar_test_data = cifar_test_data_dict[b'data']
cifar_test_labels = []
cifar_test_labels += cifar_test_data_dict[b'labels']

KNearest = 25 

cifar_test_data = cifar_test_data.reshape((len(cifar_test_data), 3, 32, 32))
cifar_test_data = np.rollaxis(cifar_test_data, 1, 4)

rgb_weights = [0.2125 , 0.5870, 0.1140]
cifar_test_data = np.dot(cifar_test_data[...,:3], rgb_weights)

counter =0
for i in range (cifar_test_data.shape[0]):
    prediction = predictClassKNN(cifar_test_data[i],cifar_train_data, cifar_train_data.shape[0], cifar_train_labels, KNearest)
    if prediction == cifar_test_labels[i]:
        counter+=1

Accuracy = (counter/cifar_test_data.shape[0]) * 100
print ("# Task Four: Finding accuracy using test data (10000) comparing with train data (50000)")
print ("Accuracy: ",Accuracy+"%")
```
